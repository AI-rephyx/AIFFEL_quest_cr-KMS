```python
import 'package:flutter/material.dart';

void main() {
  runApp(MyApp());
}

class MyApp extends StatelessWidget {
  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      home: FirstPage(),
    );
  }
}

class FirstPage extends StatefulWidget {
  @override
  _FirstPageState createState() => _FirstPageState();
}

class _FirstPageState extends State<FirstPage> {
  bool isCat = true; // 초기값은 true (고양이)

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(
        leading: Icon(Icons.pets), // 고양이 아이콘
        title: Text("First Page"),
      ),
      body: Center(
        child: Column(
          mainAxisAlignment: MainAxisAlignment.end,
          crossAxisAlignment: CrossAxisAlignment.center,
          children: [
            ElevatedButton(
              onPressed: () async {
                // Next 버튼 눌렀을 때 isCat을 false로 변경하여 SecondPage로 넘기기
                bool result = await Navigator.push(
                  context,
                  MaterialPageRoute(
                    builder: (context) => SecondPage(isCat: false), // SecondPage로 이동할 때 isCat을 false로 전달
                  ),
                );

                // SecondPage에서 반환된 값을 받아서 isCat 값 업데이트 (이 경우에는 필요 없음)
                setState(() {
                  isCat = result; // SecondPage에서 반환된 값으로 상태 업데이트
                });
              },
              child: Text("Next"),
            ),
            SizedBox(height: 30),
            GestureDetector(
              onTap: () {
                // 이미지 탭 시 isCat 값 출력
                print("isCat: $isCat");
              },
              child: Image.asset(
                'assets/cat.png', // 고양이 이미지
                width: 300,
                height: 300,
              ),
            ),
          ],
        ),
      ),
    );
  }
}

class SecondPage extends StatelessWidget {
  final bool isCat;

  SecondPage({required this.isCat});

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(
        leading: Icon(Icons.pets), // 강아지 아이콘
        title: Text("Second Page"),
      ),
      body: Center(
        child: Column(
          mainAxisAlignment: MainAxisAlignment.end,
          crossAxisAlignment: CrossAxisAlignment.center,
          children: [
            ElevatedButton(
              onPressed: () {
                // Back 버튼 눌렀을 때 isCat을 true로 변경하여 FirstPage로 돌아가기
                Navigator.pop(context, true); // isCat을 true로 반환하여 FirstPage로 돌아가기
              },
              child: Text("Back"),
            ),
            SizedBox(height: 30),
            GestureDetector(
              onTap: () {
                // SecondPage에서 isCat 값 출력
                print("isCat: $isCat"); // SecondPage에서 isCat 값 출력
              },
              child: Image.asset(
                'assets/dog.png', // 강아지 이미지
                width: 300,
                height: 300,
              ),
            ),
          ],
        ),
      ),
    );
  }
}
```
