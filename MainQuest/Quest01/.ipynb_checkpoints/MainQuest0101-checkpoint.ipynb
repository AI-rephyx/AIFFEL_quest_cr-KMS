```python
import random # 랜덤 모듈 임포트해오기

class Account: # Account 클래스 생성
    account_count = 0

    def __init__(self, account_name, balance):
        self.bank_name = "SC은행"  # 은행 이름 설정
        self.account_name = account_name  # 예금주 이름 설정
        self.account_number = self.generate_account_number()  # 계좌번호 생성
        self.balance = balance if balance > 0 else 0  # 초기 잔액 설정
        self.deposit_count = 0  # 입금 횟수 초기화
        self.deposit_history = []  # 입금 내역 리스트
        self.withdraw_history = []  # 출금 내역 리스트
        Account.account_count += 1  # 계좌 생성 시 클래스 변수 증가

    def generate_account_number(self):
        # 계좌번호를 XXX-YY-ZZZZZZ 형식으로 랜덤 생성
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

    @classmethod
        # 생성된 계좌의 개수를 반환하는 메서드
    def get_account_num(cls):
        return cls.account_count

        # 입금 메서드와 이자 지급
    def deposit(self, amount):
        if amount < 1:
            print("입금 금액은 1원 이상이어야 합니다.")
            return
        self.balance += amount  # 잔액 증가
        self.deposit_count += 1  # 입금 횟수 증가
        self.deposit_history.append(amount)  # 입금 내역 저장
        if self.deposit_count % 5 == 0:
            interest = self.balance * 0.01  # 1% 이자 계산
            self.balance += interest  # 이자 추가
            print(f"5회 입금 하셨으므로, 1% 이자가 추가되었습니다. 이자: {interest:.0f}원")

        # 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액 부족, 출금할 수 없습니다.")
            return
        self.balance -= amount  # 잔액 감소
        self.withdraw_history.append(amount)  # 출금 내역 저장

        # 계좌 정보 출력 메서드
    def display_info(self):
        formatted_balance = f"{self.balance:,.0f}원"
        print(f"은행이름 : {self.bank_name}, 예금주 : {self.account_name}")
        print(f"계좌번호 : {self.account_number}, 잔고 : {formatted_balance}")

        # 입금 내역 출력 메서드
    def show_deposit_history(self):
        print(f"{self.account_name}님의 입금 내역:")
        for idx, amount in enumerate(self.deposit_history, 1):
            print(f"{idx}번째 입금: {amount}원")

        # 출금 내역 출력 메서드
    def show_withdraw_history(self):
        print(f"{self.account_name}님의 출금 내역:")
        for idx, amount in enumerate(self.withdraw_history, 1):
            print(f"{idx}번째 출금: {amount}원")

# Q8. 여러 객체 생성
account1 = Account("모두연", 20000) # 계좌 이름과 초기 자금 설정
account2 = Account("아이펠", 2500000)
account3 = Account("캠퍼스", 3500000)
accounts = [account1, account2, account3]

# Q4. 입금 예시와 이자 지급
account1.deposit(5000)
account1.deposit(4000)
account1.deposit(3000)
account1.deposit(2000)
account1.deposit(1000)  # 이자 지급 발생

# Q5. 출금 예시
account2.withdraw(500000)
account3.withdraw(1000000)

# Q6. 정보 출력 예시
for acc in accounts:
    acc.display_info()

# Q9. 잔고가 100만원 이상인 고객의 정보만 출력
print("\n잔고가 1,000,000원 이상인 계좌 정보:")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

# Q10. 입금과 출금 내역 출력
print("\n입금 내역:")
for acc in accounts:
    acc.show_deposit_history()

print("\n출금 내역:")
for acc in accounts:
    acc.show_withdraw_history()

# Q3. 클래스 변수 출력
print(f"\n전체 계좌 개수: {Account.get_account_num()}개")
```
